Where in numpy

```python
arr = np.array([10,-3,4,0, 6,-9])
result = np.where(arr< 0, 0 , arr) #if condition true then zero, else output the array as it is
print(result)
```

```python
arr= np.array([1,2,3,5,6])
result = np.where(arr % 2 != 0, -1, arr)
print(result)
```

```python
marks = np.array([23,45,67, 30, 87])
status = np.where(marks>50, 1, 0)
print(status)
```

```python
data = np.array([10,20,50,70,70])
mean_val = np.mean(data)
result = np.where(data > mean_val, mean_val, data)
print(result)
```

```python
mat = np.array([[10,80,30],[40,50,60]])
result= np.where(mat > 50, 1, 0)
print(result)
```
